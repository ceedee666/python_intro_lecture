# Aufgabe 5
In einer Bank sollen Kreditkartenanträge automatisch bearbeitet werden. Sie sollen nun das Regelwerk zur Entscheidung über
Anträge einem Python Programm umsetzen.
Folgende Regeln sollen geprüft werden:
- Falls der Antragsteller jünger als 18 Jahre alt ist wird keine Kreditkarte ausgestellt.
- Falls der Antragsteller zwischen 18 und 30 Jahre alt ist wird nur eine Kreditkarte ausgestellt, wenn das Einkommen über 1500€ pro Monat liegt
  - Ist der Antragsteller zwischen 18 und 30 und sein Einkommen liegt über 1500€ pro Monat wird eine Kreditkarte mit einem Limit von 2000€ ausgestellt.
  - Ist der Antragsteller zwischen 18 und 30 und sein Einkommen liegt über 3000€ pro Monat wird eine Kreditkarte mit einem Limit von 4000€ ausgestellt.
- Ist der Antragsteller zwischen 31 und 65 und sein Einkommen liegt unter 2000€ pro Monat wird eine Kreditkarte mit einem Limit von 3000€ ausgestellt.
- Ist der Antragsteller zwischen 31 und 65 und sein Einkommen liegt über 2000€ pro Monat wird eine Kreditkarte mit einem Limit von 6000€ ausgestellt.
- Ist der Antragsteller über 65 wird nur eine Kreditkarte ausgestellt falls sein Einkommen über 2000€ pro Monat liegt.
  - In diesem Fall mit einem Kreditlimit von 3000€.

Implementieren Sie nun die Funktion `antrag_pruefen`. Diese Funktion erhält zweieinen Parameter
- `monatliches_einkommen` vom Typ `Integer` und enthält das monatliche Einkommen des Antragstellers
- `alter` vom Typ `Integer` und enthält das Alter des Antragstellers.

Als Ergebnis soll die Funktion die Variable `limit` berechnen, die das Limit der Kreditkarte enthält.

```python
def antrag_pruefen(montaliches_einkommen, alter):
    ### START

    if #Bedingung Alter ist kleiner als 18:
        limit = 0
    elif #Komplete Bedingung Alter kleier oder gleich 30 and montaliches Einkommen > 1500:
        if #Bedingung monatliches Einkommen ist kleiner oder gleich 3000:
            limit = 2000
        else:
            limit = 4000
    elif #Komplete Bedingung Alter ist zwischen 31 und 65:
        if #Bedingung monatliches Einkommen ist kleiner oder gleich 2000:
            limit = 3000
        else:
            limit = 6000
    elif #Bedingung monatliches Einkommen ist größer 2000:
        limit = 3000
    else:
        limit = 0
          
    ### ENDE
    return limit

print("Wenn Ihre Funktion mit den Parametern \"montaliches_einkommen = 500\" und \"alter = 25\" aufgerufen wird enthält die Variable `limit` den Wert", antrag_pruefen(500, 25))
```
