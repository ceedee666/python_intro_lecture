Zur Bewertung des Risikos sollen folgende Regeln verwendet werden:
- Falls das Alter eines Hauses kleiner oder gleich 10 Jahre ist und es sich um ein Reihenhaus oder Einfamilienhaus handelt, ist die Risikoklasse A
- Falls es sich um ein Reihenhaus in Alter von 11 - 20 Jahren handelt ist die Risikoklasse B

- Bei Häusern ab einem Alter von 21 Jahren ist die Risikoklasse D.

Implementieren Sie nun die Funktion `risikobewertung`. Diese Funktion erhält folgende Parameter:
- Der Parameter `alter` gibt das Alter des Gebäudes in Jahren an und ist vom Typ `Integer`.
- Der Parameter `gebäudetyp` gibt an ob das Gebäude ein Einfamilienhaus, Reihenhaus oder Mehrfamilienhaus ist und ist vom Typ `String`.
     - Ist das Gebäude ein Einfamilienhaus enthält der Parameter `gebaeudetyp` den Wert `"EFH"`
     - Ist das Gebäude ein Reihenhaus enthält der Parameter `gebaeudetyp` den Wert `"RH"`
     - Ist das Gebäude ein Mehrfamilienhaus enthält der Parameter `gebaeudetyp` den Wert `"MFH"`

Als Ergebnis der Funktion `risikobewertung` sollen die Variable `risikoklasse` berechnet werden.
Falls Parameter `gebäudetyp` einen ungültigen Wert enthält oder nicht vom typ String ist sollen die Variable `fehlermeldung`  auf den Wert `"Parameterfehler"` gesetzt werden.
