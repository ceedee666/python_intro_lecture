 # Aufgabe 1

 In einer Bank sollen Kreditkartenanträge automatisch bearbeitet werden. Sie sollen nun das Regelwerk zur Entscheidung über
 Anträge einem Python Programm umsetzen.
 In ersten Schritt soll überprüft werden ob ein Antragsteller jünger als 18 Jahren alt ist. In diesem Fall soll keine
 Kreditkarte ausgestellt werden. Alle Antragsteller ab 18 Jahren sind berechtigt eine Kreditkarte zu erhalten.

 Implementieren Sie nun die Funktion `antrag_pruefen`. Diese Funktion erhält einen Parameter `alter`,
 welcher in Form eines `Integer`-Wertes das Alter des Antragstellers enthält.
 Als Ergebnis soll die Variable `ist_berechtigt` berechnet werden. Ist der Antragsteller älter als 18 Jahre
 soll die Varibale `ist_berechtigt` auf den Wert `True` gesetzt werden, ansonsten auf den Weret `False#`.

 ## Wichtige Hinweise
 Bisher haben Sie im Rahmen dieser Einführung in Python Funktionen noch nicht kennen gelernt. Aus diesem Grund hier ein paar Hinweise zum Umgang mit Funktionen im Rahmen dieser Übungsaufgaben:

 ### Hilfe
 Sie können Ihre Lösung mit den automatisierten [Tests](#Automatische-Tests) weiter unten im Dokument überprüfen.<br>
 Eine Einführung in die `Kontrollstrukturen`in Python finden Sie [hier](../../30_conditionals/10_conditionals.ipynb).

```python
def antrag_pruefen(alter):
    ### START
    ist_berechtigt = True
    if alter < 18:
        ist_berechtigt = False
        
    ### ENDE
    return ist_berechtigt

print("Wenn Ihre Funktion mit dem Parameter \"alter = 25\" aufgerufen wird ist das Ergebnis:", antrag_pruefen(25))
```
